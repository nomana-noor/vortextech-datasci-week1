```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('amazon.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['rating_count'])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df[['discounted_price', 'actual_price', 'rating', 'discount_percentage']].head()
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False).astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False).astype(float)
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '', regex=False).astype(float)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df.info()
```

```python
df = df.dropna(subset=['rating'])
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['rating_count'].head()
```

```python
df['rating_count'] = df['rating_count'].str.replace(',', '', regex=False).astype(float)
```

```python
df.info()
```

```python
df['category'].head()
```

```python
df['main_category'] = df['category'].str.split('|').str[0]
```

```python
plt.figure(figsize=(12,7))
df['main_category'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Categories by Number of Products')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Insight:
This chart shows that Electronics products dominate the dataset, with the top category having significantly more products than others. The distribution is skewed, with a few categories making up most of the listings.

```python
plt.figure(figsize=(10,6))
df['rating'].plot(kind='hist', bins=20, color='coral', edgecolor='black')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency (Number of Products)')
plt.show()
```

### Insight:
Most products have a rating between 4.0 and 4.5, showing that customers generally rate products positively. Very few products have ratings below 3.0.

```python
top5 = df['main_category'].value_counts().head(5)

def autopct_format(pct):
    return f'{pct:.1f}%' if pct > 2 else ''

plt.figure(figsize=(8,8))
plt.pie(top5, labels=None, autopct=autopct_format, startangle=90, colors=sns.color_palette('pastel'))
plt.title('Top 5 Categories Share of Products')
plt.legend(top5.index, title="Category", loc="best", bbox_to_anchor=(1, 0.5))
plt.show()
```

### Insight:
Electronics, Computers & Accessories, and Home & Kitchen together make up over 95% of all products, showing the dataset is heavily focused on these three categories.

```python
plt.figure(figsize=(10,6))
df['discounted_price'].plot(kind='hist', bins=30, color='mediumseagreen', edgecolor='black')
plt.title('Distribution of Discounted Prices')
plt.xlabel('Discounted Price (₹)')
plt.ylabel('Frequency (Number of Products)')
plt.show()
```

### Insight:
The majority of products are priced under ₹5,000, with very few products priced above ₹20,000. This shows the dataset mostly contains affordable, everyday products.

```python
plt.figure(figsize=(10,6))
plt.scatter(df['discount_percentage'], df['rating'], alpha=0.5, color='purple')
plt.title('Discount Percentage vs Rating')
plt.xlabel('Discount Percentage (%)')
plt.ylabel('Rating')
plt.show()
```

### Insight:
There is no strong relationship between discount percentage and rating. Ratings stay mostly between 3.5 and 4.5 regardless of how much discount is offered, suggesting discount level doesn't significantly affect customer satisfaction.
